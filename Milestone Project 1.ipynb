```python
#### MILESTONE PROJECT 1 ######
from IPython.display import clear_output
def display_board(board):
    print(board[7]+'|'+board[8]+"|"+board[9])
    print(board[4]+'|'+board[5]+"|"+board[6])
    print(board[1]+'|'+board[2]+"|"+board[3])
```

```python
def player_unit():
    player1 = ""
    while player1.lower() not in ('x','o'):
        player1 = input("Seleccione con cual letra desea jugar. ¿X u O?")
    return player1.upper()
```

```python
def place_maker(board,maker,position):
    if board[position] not in ('X','O'):
        board[position] = maker
    return board
```

```python
def win_check(board,mark):
    if board[1] == board [2] == board[3] and board[1] == mark:
        return True
    elif board[4] == board[5] == board[6] and board[4] == mark:
        return True
    elif board[7] == board[8] == board[9] and board[7] == mark:  
        return True
    elif board[1] == board[5] == board[9] and board[1] == mark:
        return True
    elif board[3] == board[5] == board[7] and board[3] == mark:
        return True
    elif board[1] == board[4] == board[7] and board[1] == mark:
        return True
    elif board[2] == board[5] == board[8] and board[2] == mark:
        return True
    elif board[3] == board[6] == board[9] and board[3] == mark:
        return True
    else:
        return False
```

```python
import random 
def choose_first():
    x = random.randint(1,2)
    if(x == 1):
        return "Jugador 1"
    else:
        return "Jugador 2"
```

```python
def space_check(board, position):
    return board[position] not in ('X','O')
```

```python
def full_board_check(board):
    contador = 0
    for x in board:
        if x in ('X','O'):
            contador += 1
    return contador == 9
```

```python
def player_choice(board):
    position = input("Seleccione la posición de la marca:")
    espacio = space_check(board,int(position))
    if espacio:
        return position
    else:
        return 0
```

```python
def replay():
    x = ""
    while x not in ("1","2"):
        x = input("¿Desea volver a jugar? 1= Si. 2= No")
    return x == "1"
```

```python
def inicializar():
    test_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    return test_board
```

```python
print("Welcome to Tic Tac Toe")
x = True
z = True

while x == True:
    
    board = inicializar()
    marcaJugador1 = player_unit()
    fullboard = False
    chequeo = False
    if marcaJugador1 == "X":
        marcaJugador2 = "O"
    else:
        marcaJugador2 = "X"
    
    juegoJugador1 = True
    #choose_first()
   
    while fullboard == False:
        while juegoJugador1 == True and chequeo == False:
            print("---- TURNO JUGADOR 1 ----")
            position = player_choice(board)
        
            while position == 0:
                print("Campo ya ha sido usado, elegir otro")
                position = player_choice(board)
        
            place_maker(board,marcaJugador1,int(position))
            display_board(board)
            chequeo = win_check(board,marcaJugador1)
        
            if(chequeo):
                print("***FELICIDADES*** Ganador Jugador1")
                chequeo = True
                fullboard = True
                x = replay()
            else:
                fullboard = full_board_check(board)
        
            if(fullboard == True and chequeo == False):
                print("------EMPATE-----")
                chequeo = True
                fullboard = True
                x = replay()
            
            juegoJugador1 = False
    
        while juegoJugador1 == False and chequeo == False:
            print("---- TURNO JUGADOR 2 ----")
            position = player_choice(board)
        
            while position == 0:
                print("Campo ya ha sido usado, elegir otro")
                position = player_choice(board)
        
            place_maker(board,marcaJugador2,int(position))
            display_board(board)
            juegoJugador1 = True
        
            chequeo = win_check(board,marcaJugador2)
        
            if(chequeo):
                print("***FELICIDADES*** Ganador Jugador2")
                chequeo = True
                fullboard = True
                x = replay()
            else:
                fullboard = full_board_check(board)
                
            if(fullboard == True and chequeo == False):
                print("------EMPATE-----")
                chequeo = True
                fullboard = True
                x = replay()    

print("Gracias por jugar")
    
    
```
